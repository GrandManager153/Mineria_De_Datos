####Conceptos

La minería de datos es una técnica asistida por computadora que se utiliza en los análisis para procesar y explorar grandes conjuntos de datos. Gracias a las herramientas y métodos de minería de datos, las organizaciones pueden descubrir patrones y relaciones ocultas en sus datos. La minería de datos transforma datos en bruto en conocimiento práctico. Las compañías utilizan dicho conocimiento para resolver problemas, analizar las consecuencias en el futuro de decisiones empresariales y aumentar sus márgenes de beneficio.

Es un término erróneo porque el objetivo de esta actividad no es extraer o minar los datos en sí. Más bien, ya existe una gran cantidad de datos, y la minería de datos extrae un significado o un conocimiento valioso de estos.

Understanding Data Types in Python

```python
# Importa numpy y lo renombra como np
import numpy as np
```

```python
# Arreglo lineal de enteros
np.array([1, 4, 2, 5, 3])
```

```python
# Castea(cambia) la lista al primer valor que venga en ella
np.array([3.14, 4, 2, 3])
```

```python
# Se puede definir(implicitar) el tipo de datos en los que se quiere la lista
np.array([1, 2, 3, 4], dtype='float32')
```

```python
# Cuando el argumento de np.array es una iteracion anidada genera una matriz(lista de listas)
np.array([range(i, i + 3) for i in [2, 4, 6]])
```

```python
# Forma de impresion en python(lista de listas)
[range(i, i + 3) for i in [2, 4, 6]]
```

Creating Arrays from Scratch

```python
# Forma de crear un arreglo lleno de 0
np.zeros(10, dtype=int)
```

```python
# Forma de crear un arreglo Bidimensional
np.ones((3, 5), dtype=float)
```

```python
# Forma de crear un arreglo bidimensional y que se llene de un solo dato
np.full((3, 5), 3.14)
```

```python
# Forma de crear un arreglo con una secuencia en este caso de cada 2
np.arange(0, 20, 2)
```

```python
# Forma de crear un arreglo con 5 valores entre un rango
np.linspace(0, 1, 5)
```

```python
# Forma de crear un arreglo sobre la distribucion estandar
np.random.random((3, 3))
```

```python
# Forma de crear un arreglo de datos aleatorios entre 0 y 1 sobre la forma de distribucion estandar
np.random.normal(0, 1, (3, 3))
```

```python
# Forma de crear un arreglo aleatorios de enteros en un intervalo entre entre 0 y 10
np.random.randint(0, 6, (3, 3))
```

```python
# Forma de crear una matriz como una matriz identidad
np.eye(3)
```

```python
# Forma de crear un arreglo sin inicializar con enteros
np.empty(3)
```

The Basics of NumPy Arrays

```python
# Creacion de diversos arreglos
np.random.seed(0)  # seed for reproducibility

x1 = np.random.randint(10, size=6)  # One-dimensional array
x2 = np.random.randint(10, size=(3, 4))  # Two-dimensional array
x3 = np.random.randint(10, size=(3, 4, 5))  # Three-dimensional array
```

```python
# Caracteristicas de los valores dentro del arreglo
print("x3 ndim: ", x3.ndim)
print("x3 shape:", x3.shape)
print("x3 size: ", x3.size)
```

```python
# Tipo de datos en el arreglo
print("dtype:", x3.dtype)
```

```python
# Tamaño de los datos en el arreglo y cuanto ocupan
print("itemsize:", x3.itemsize, "bytes")
print("nbytes:", x3.nbytes, "bytes")
```

Media

```python
data = np.array((1, 2, 3, 4, 5))
weight = np.array((1, 2, 3, 4, 5))
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weight)

print(promedio)
print(promedio_ponderado)
```

Mediana

```python
mediana = np.median(data)
print(mediana)
```

Varianza y Desviacion Estandar

```python
varianza = np.var(data)
desviacion_estandar = np.std(data)

print(varianza)
print(desviacion_estandar)
```

Generacion de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000)
print(np.mean(muestra), np.std(muestra))
```

Correlacion

```python
corr = np.corrcoef((1, 2, 3), (1, 5, 7))
print(corr)
```
